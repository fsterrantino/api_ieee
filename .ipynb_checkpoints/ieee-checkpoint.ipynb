```python
def create_url(start_record):
    url = 'https://ieeexploreapi.ieee.org/api/v1/search/articles?parameter&apikey='
    apiKey = 'bggc8khstkgmzsc7myse9yfc'

    maxRecords = 10
    basicQuery = f'&content_type=Journals&max_records={maxRecords}'

    
    queryText = '&querytext=(gui OR ui OR user interface)'
    #                AND
    #                (sketch OR image OR mock up OR mock-up OR prototype OR wireframe OR screenshot)
    #                AND
    #                ("automatic code generation" OR "automatic programming" OR "automatic code" OR "automatic software development" OR "to code")
    #            
    
    searchYear = 2023
    queryYear = f'&start_year={searchYear}'

    startRecord = start_record
    queryStartRecord = f'&start_record={startRecord}'
    
    queryUrl = url + apiKey + basicQuery + queryText + queryYear + queryStartRecord
    return queryUrl
```
